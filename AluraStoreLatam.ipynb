### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_tienda = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"Las ventas de la tienda 1 suman un total de {ingresos_tienda}")
print(f"Las ventas de la tienda 2 suman un total de {ingresos_tienda2}")
print(f"Las ventas de la tienda 3 suman un total de {ingresos_tienda3}")
print(f"Las ventas de la tienda 4 suman un total de {ingresos_tienda4}")
```

```python
plt.pie([ingresos_tienda, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4],labels=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], autopct="%1.1f%%")
plt.title("Porcentaje de ventas por tienda")
```

# 2. Ventas por categoría

```python
categorias1 = tienda['Categoría del Producto'].value_counts()
categorias2 = tienda2['Categoría del Producto'].value_counts()
categorias3 = tienda3['Categoría del Producto'].value_counts()
categorias4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print(f"Estas son las ventas más populares, por categorías en la tienda 1: \n{categorias1}")
print(f"Estas son las ventas más populares, por categorías en la tienda 2: \n{categorias2}")
print(f"Estas son las ventas más populares, por categorías en la tienda 3: \n{categorias3}")
print(f"Estas son las ventas más populares, por categorías en la tienda 4: \n{categorias4}")
```

```python
plt.barh(categorias1.index, categorias1.values)
plt.title("Ventas por Categoría en tíenda 1")
```

```python
plt.barh(categorias2.index, categorias2.values)
plt.title("Ventas por Categoría en tíenda 2")
```

```python
plt.barh(categorias3.index, categorias3.values)
plt.title("Ventas por Categoría en tíenda 3")
```

```python
plt.barh(categorias4.index, categorias4.values)
plt.title("Ventas por Categoría en tíenda 4")
```

# 3. Calificación promedio de la tienda


```python
promedio1 = tienda['Calificación'].mean().round(2)
promedio2 = tienda2['Calificación'].mean().round(2)
promedio3 = tienda3['Calificación'].mean().round(2)
promedio4 = tienda4['Calificación'].mean().round(2)
```

```python
print(f"El promedio de calificación de compradores a la tienda 1 es de {promedio1}")
print(f"El promedio de calificación de compradores a la tienda 2 es de {promedio2}")
print(f"El promedio de calificación de compradores a la tienda 3 es de {promedio3}")
print(f"El promedio de calificación de compradores a la tienda 4 es de {promedio4}")
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
promedios = [promedio1, promedio2, promedio3, promedio4]
plt.plot(tiendas, promedios, marker="o")
plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")
```

```python
print(promedio1, promedio2, promedio3, promedio4)
```
